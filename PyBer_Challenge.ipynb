# Pyber Challenge

### Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib widget
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Load CSVs
city_data_df = pd.read_csv("Resources/city_data.csv")
ride_data_df = pd.read_csv("Resources/ride_data.csv")
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"], copy=False)

# Display the data table for preview
pyber_data_df.head(20)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
type_ride_count = pyber_data_df.groupby(['type']).count()['ride_id']
```

```python
# 2. Get the total drivers for each city type
type_driver_count = city_data_df.groupby(['type']).sum()['driver_count']
```

```python
#  3. Get the total amount of fares for each city type
type_fare_total = pyber_data_df.groupby(['type']).sum()['fare']
```

```python
#  4. Get the average fare per ride for each city type. 
type_fare_avg = pyber_data_df.groupby(['type']).sum()["fare"] / pyber_data_df.groupby(['type']).count()['ride_id']
```

```python
# 5. Get the average fare per driver for each city type. 
type_driver_avg = pyber_data_df.groupby(['type']).sum()["fare"] / city_data_df.groupby(['type']).sum()['driver_count']
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({'Total Rides':type_ride_count,
                                'Total Drivers':type_driver_count,
                                'Total Fares':type_fare_total,
                                'Average Fare per Ride':type_fare_avg,
                                'Average Fare per Driver':type_driver_avg})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df.dtypes
```

```python
#  8. Format the columns.
pyber_summary_df.style.format({'Total Rides':'{:20,.0f}',
                               'Total Drivers':'{:20,.0f}',
                               'Total Fares': '${:20,.2f}',
                              'Average Fare per Ride': '${:20,.2f}',
                              'Average Fare per Driver': '${:20,.2f}'
                              })
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# DataFrame showing the sum of the fares for each date where the indices are the city type and date.
fare_by_date_df = pyber_data_df.groupby(['type','date']).sum()['fare']
```

```python
# Reset the index on the DataFrame.
fare_by_date_df = fare_by_date_df.reset_index()
```

```python
# Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
pivot_fare = fare_by_date_df.pivot(index='date', columns='type', values='fare')
pivot_fare.head(10)
```

```python
# Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_to_may_df = pivot_fare.loc['2019-01-01':'2019-04-29']
jan_to_may_df
```

```python
# Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
jan_to_may_df.index = pd.to_datetime(jan_to_may_df.index)
```

```python
# Check that the datatype for the index is datetime using df.info()
jan_to_may_df.info()
```

```python
# Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_to_may_resampled_df = jan_to_may_df.resample('W').sum()
```

```python
jan_to_may_resampled_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function.
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
jan_to_may_resampled_df.plot(figsize=(30,12))
plt.title('Total Fare by City Type')
plt.ylabel('Fare ($USD)')
plt.legend(loc='center')
```
