```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
```

```python
# Create the plot.
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3, marker="*", color="g", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(-5, 59)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(8,8))
```

```python
# Create the plot.
ax.barh(x_axis, y_axis, color="darkturquoise", label='Boston')
# Set major and minor ticks
ax.set_xticks(np.arange(0,51,step=5.0))
ax.xaxis.set_minor_locator(MultipleLocator(1))
# Create labels for the x and y axes.
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()
# Invert Axis
ax.invert_yaxis()
```

```python
fig, ax = plt.subplots()
```

```python
# Create the Scatter
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], alpha=0.8, color='skyblue', label='Boston', linewidths=2, edgecolor='k')
# Set the limits
ax.set_xlim(0,50)
# Label the axes
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Create the Title
ax.set_title("PyBer Fare by Month")
# Add the Legend
ax.legend()
# Invert Axes
ax.invert_yaxis()
```

```python
# Assign 12 colors, one for each month. Set explode values.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
      counterclock=False,
      startangle=90,
      shadow=True)
```
