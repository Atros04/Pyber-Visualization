```python
#import dependenceis
import matplotlib.pyplot as plt
import statistics
import numpy as np
from matplotlib.ticker import MultipleLocator
import pandas as pd
```

```python
#Read the CSV
rides_df = pd.read_csv("Resources/PyBer_ride_data.csv")
rides_df
```

```python
# plot x axis
x_axis = np.arange(len(rides_df))
tick_locations = [value for value in x_axis]
stdev = statistics.stdev(rides_df['Avg. Fare ($USD)'])

rides_df.plot(x='Month',y='Avg. Fare ($USD)',kind='bar', yerr=stdev, capsize=3, color='skyblue')
plt.xticks(tick_locations, rides_df["Month"], rotation=0)
plt.show()
```
